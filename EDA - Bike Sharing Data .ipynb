```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Bike Sharing dataset
file_path = r'C:\Users\Khushi Bakshi\Downloads\bike+sharing+dataset.zip'
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the data has been loaded correctly
df.head()
```

```python
# Display summary statistics
df.describe()
```

```python
# Visualize the distribution of bike rentals over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='dteday', y='cnt', data=df, color='blue')
plt.title('Bike Rentals Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Bike Rentals')
plt.show()
```

```python
# Explore the impact of weather on bike rentals
plt.figure(figsize=(10, 6))
sns.boxplot(x='weathersit', y='cnt', data=df, palette='viridis')
plt.title('Impact of Weather on Bike Rentals')
plt.xlabel('Weather Situation')
plt.ylabel('Number of Bike Rentals')
plt.show()
```

```python
# Analyze the correlation between numerical variables
correlation_matrix = df[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
